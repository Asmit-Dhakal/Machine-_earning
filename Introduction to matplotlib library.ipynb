# Making figure with NumPy array

  

```python
# Creat e some data
x=np.linspace(0,10,100)
x
x[:10]
```

```python
# plot the data scalar plots
```

```python
fig, ax = plt.subplots()
ax.plot(x,x**2); # add some data
type(fig), type(ax)
```

```python
# Use some data to make a scatterZ
fig,ax=plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
# Use some data to make a scatter
fig,ax=plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
# Make a plot from dictionary
Nut={"A":10,
     "B":6,
     "C":20
    }
fig,ax=plt.subplots()
ax.bar(Nut.keys(),Nut.values())
ax.set(title="Store",ylabel="Price($)");
```

```python
fig,ax=plt.subplots()
ax.barh(list(Nut.keys()),list(Nut.values()))
```

```python
# Make some data
x=np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x);
```

```python
# two option for subplots
```

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(x,x/20);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(Nut.keys(),Nut.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplot 2
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
# Plot to each diffrent axis
ax[0,0].plot(x,x/20);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(Nut.keys(),Nut.values())
ax[1,1].hist(np.random.randn(1000));
```
